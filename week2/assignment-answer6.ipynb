```python
def answer_six():
    census_df = pd.read_csv('census.csv') 
    county_df = census_df[census_df['SUMLEV'] == 50]
    county_df = county_df[['STNAME', 'CTYNAME', 'CENSUS2010POP']]
    for group, frame in county_df.groupby(['STNAME']):
        print('Group', group, 'Frame', frame)
```
